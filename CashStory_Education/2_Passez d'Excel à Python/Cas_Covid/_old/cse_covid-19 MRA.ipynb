1.3 - Méthode 2 : Effectuer votre première boucle 

```python
# Lire les données brutes sur Github
url_init = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'

url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/'

# Liste des fichiers à récupérer sur le répertoire
files = [
         'time_series_covid19_confirmed_global.csv',
         'time_series_covid19_deaths_global.csv',
         'time_series_covid19_recovered_global.csv'
        ]

# Boucle pour récupérer et concaténer les fichier sources
# Initialisation du dataframe
data_source = pd.DataFrame()

# Boucle pour récupérer les fichier
for file in files:
    # Variabiliser la lecture des fichiers 
    df = pd.read_csv(f'{url}{file}')
    # Créer une nouvelle colonne pour distinguer les fichiers
    df['FILENAME'] = file
    # Concaténation des fichiers  
    data_source = pd.concat([data_source, df], axis=0)

# Visualiser votre dataframe 
data_source.head(5)
```

#### Etape 2 - Nettoyer votre base de donnée

2.1 - Conserver les colonnes qui vous intéresse

```python
# Méthode 1 : Conserver les colonnes qui vous intéresse 
# Initialisation du dataframe
data_clean = data_source.copy()

# Suppression des colonnes inutiles
data_clean = data_clean.drop(['Province/State', 'Lat', 'Long'], axis=1)

# Visualiser votre dataframe  
data_clean.head(5)
```

```python
# Méthode 2 : Conserver les colonnes qui vous intéresse 
# Initialisation du dataframe
data_clean = data_source.copy()

# Colonnes à supprimer
to_drop = ['Province/State', 'Lat', 'Long']

# Suppression des colonnes inutiles
data_clean = data_clean.drop(to_drop, axis=1)

# Visualiser votre dataframe  
data_clean.head(5)
```

2.2 - Renommer les lignes

```python
# Sélectionner une suite de caractère
data_clean['TYPE'] = data_clean['FILENAME'].str[20:-11]

# Mettre en majuscule le premier caractère
data_clean['TYPE'] = data_clean['TYPE'].str.capitalize()

# Visualiser votre dataframe  
data_clean.head(5)
```

2.3 -  Modifier les en-têtes de colonne

```python
# Méthode 1 : Renommer la colonne
# Initialisation du dataframe
data_clean = data_clean.copy()

# Renommer la colonne
data_clean = data_clean.rename(columns={'Country/Region': 'Country'})

# Visualiser votre dataframe 
data_clean.head(5)
```

```python
# Méthode 2 : Renommer la colonne
# Initialisation du dataframe
data_clean = data_clean.copy()

# Renommer la colonne
to_rename = {'Country/Region': 'Country'}
data_clean = data_clean.rename(columns=to_rename)

# Visualiser votre dataframe 
data_clean.head(5)
```

```python
# Mettre en majuscule toutes les colonnes
data_clean.columns = map(lambda x: str(x).upper(), data_clean.columns)

# Visualiser votre dataframe 
data_clean.head(5)
```

#### Etape 3 - Mise au format base de donnée

3.1 - Pivoter les colonnes en ligne

```python
# Initialisation du dataframe
data_pivot = data_clean.copy()

# Colonnes à conserver
to_keep = ['TYPE', 'COUNTRY', 'FILENAME']

# Pivot des colonnes
data_pivot = data_pivot.melt(id_vars=to_keep, 
                             value_name='VALUE', 
                             var_name='DATE')

# Visualiser votre dataframe 
data_pivot.head(5)
```

3.2 - Agréger les lignes

```python
# Colonnes à aggréger
to_group = ['DATE', 'COUNTRY', 'TYPE']

# Regroupement et aggrégation
data_pivot = data_pivot.groupby(to_group, as_index=False).agg({'VALUE': 'sum'})

#-- Convertir au format date
data_pivot['DATE'] = pd.to_datetime(data_pivot['DATE'], format='%m/%d/%y')

# Visualiser votre dataframe 
data_pivot.head(5)
```

#### Etape 4 - Enrichissement des données

4.1 - Ajout d'une règle spécifique

```python
# Initialisation du dataframe
date_max = data_pivot.copy()

# Calcul max date 
# Méthode 1 : fonction max
date_max['MAX_DATE_1'] = date_max['DATE'].max()

# Méthode 2 : fonction max avec critères
to_group = ['COUNTRY']
date_max['MAX_DATE_2'] = date_max.groupby(to_group)["DATE"].transform('max')

# Conditions : Si c'est la dernière date du pays alors "1" sinon "0"
last_date = date_max.copy()
# Méthode 1 : fonction apply
last_date['LAST_DATE_1'] = last_date.apply(lambda row: 1 if row['DATE'] == row['MAX_DATE_2'] else 0, axis=1)

# Méthode 1 : fonction loc
last_date.loc[last_date['DATE'] == last_date['MAX_DATE_2'], 'LAST_DATE_2'] = 1
last_date.loc[last_date['DATE'] != last_date['MAX_DATE_2'], 'LAST_DATE_2'] = 0

last_date
```

4.2 - Ajouter des continents à la base de donnée

```python
# Initialisation du dataframe
data_enr = last_date.copy()

# Suppression des colonnes
to_drop = ['MAX_DATE_1', 'LAST_DATE_1']
data_enr = data_enr.drop(to_drop, axis=1)

# Utilisation d'un référentiel pour enrichir la base de donnée
# Lecture du fichier REF_CONTINENT
ref_continent = pd.read_excel('REF_CS.xlsx',
                              sheet_name='REF_CONTINENT')

# Jointure avec le dataframe principal
data_enr = pd.merge(data_enr, 
                    ref_continent, 
                    on=['COUNTRY'],
                    how='left'
                   )

# Affecter une valeur au pays qui non pas de continent 
# Remplacer le "NaN" par 0
data_enr['CONTINENT'] = data_enr['CONTINENT'].fillna(0)
# Remplacer les 0 par “To be affected”
data_enr.loc[data_enr['CONTINENT'] == 0, 'CONTINENT'] = 'To be affected'

data_enr.head(5)
```

4.3 - Calcul des cas actifs

```python
# Calcul des cas actifs
data_active = data_enr.copy()
data_active.loc[data_active['TYPE'] != 'Confirmed', 'VALUE'] = data_active['VALUE'] * (-1)
data_active.loc[:, 'TYPE'] = 'Active Cases'

# Colonnes à grouper
to_group = ['DATE', 'COUNTRY', 'CONTINENT', 'TYPE']

# Colonnes à aggréger 
to_agg = {'VALUE': 'sum'}

# Regroupement et aggrégation
data_active = data_active.groupby(to_group, as_index=False).agg(to_agg)

# Concaténation avec le dataframe principal
data_bdd = pd.concat([data_enr, data_active], axis=0, sort=False)

data_bdd.head(5)
```

#### Etape 5 : Sauvegarde et exposition du csv

```python
# Ajout de la date de mise à jour
data_bdd['DATE_MAJ'] = datetime.now().strftime('%Y/%m/%d %H:%M:%S')

# Sauvegarde des données au format csv
data_bdd.to_csv('BASE_COVID.csv',
                sep=";",
                decimal=",",
                index=False)

print('CSV sauvegardé dans le répertoire courant.')
```
